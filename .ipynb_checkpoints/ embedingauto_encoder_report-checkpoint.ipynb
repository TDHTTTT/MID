````python
def sample_imgs(feature, target, size):
    selected = np.random.randint(feature.shape[0]-size, size=1)[0]
    f = feature[selected:selected+size,:,:,:]
    t = target[selected:selected+size,:]
    return f,t
````

````python
class ClusteringLayer(Layer):
    """
    Clustering layer converts input sample (feature) to soft label, i.e. a vector that represents the probability of the
    sample belonging to each cluster. The probability is calculated with student's t-distribution.

    # Example
    ```
        model.add(ClusteringLayer(n_clusters=10))
    ```
    # Arguments
        n_clusters: number of clusters.
        weights: list of Numpy array with shape `(n_clusters, n_features)` witch represents the initial cluster centers.
        alpha: degrees of freedom parameter in Student's t-distribution. Default to 1.0.
    # Input shape
        2D tensor with shape: `(n_samples, n_features)`.
    # Output shape
        2D tensor with shape: `(n_samples, n_clusters)`.
    """

    def __init__(self, n_clusters, weights=None, alpha=1.0, **kwargs):
        if 'input_shape' not in kwargs and 'input_dim' in kwargs:
            kwargs['input_shape'] = (kwargs.pop('input_dim'),)
    
        super(ClusteringLayer, self).__init__(**kwargs)
        self.n_clusters = n_clusters
        self.alpha = alpha
        self.initial_weights = weights
        self.input_spec = InputSpec(ndim=2)

    def build(self, input_shape):
        assert len(input_shape) == 2
        input_dim = input_shape[1]
        self.input_spec = InputSpec(dtype=K.floatx(), shape=(None, input_dim))
        self.clusters = self.add_weight((self.n_clusters, input_dim), initializer='glorot_uniform', name='clusters')
        if self.initial_weights is not None:
            self.set_weights(self.initial_weights)
            del self.initial_weights
        self.built = True

    def call(self, inputs, **kwargs):
        """ student t-distribution, as same as used in t-SNE algorithm.
         Measure the similarity between embedded point z_i and centroid µ_j.
                 q_ij = 1/(1+dist(x_i, µ_j)^2), then normalize it.
                 q_ij can be interpreted as the probability of assigning sample i to cluster j.
                 (i.e., a soft assignment)
        Arguments:
            inputs: the variable containing data, shape=(n_samples, n_features)
        Return:
            q: student's t-distribution, or soft labels for each sample. shape=(n_samples, n_clusters)
        """

        q = 1.0 / (1.0 + (K.sum(K.square(K.expand_dims(inputs, axis=1) - self.clusters), axis=2) / self.alpha))
        q **= (self.alpha + 1.0) / 2.0
        q = K.transpose(K.transpose(q) / K.sum(q, axis=1)) # Make sure each sample's 10 values add up to 1.
        return q

    def compute_output_shape(self, input_shape):
        assert input_shape and len(input_shape) == 2
        return input_shape[0], self.n_clusters

    def get_config(self):
        config = {'n_clusters': self.n_clusters}
        base_config = super(ClusteringLayer, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))

    
````

````python
def auto_encoder(encoded_layer_size=2):
    input_img = Input(shape=(*IMAGE_SHAPE, 1))  # adapt this if using `channels_first` image data format

    x = Conv2D(IMAGE_SHAPE[0], (3, 3), activation='relu', padding='same')(input_img)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(IMAGE_SHAPE[0]//2, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    # x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
    # x = MaxPooling2D((2, 2), padding='same')(x)

    shape_before_flattening = K.int_shape(x)
    x = Flatten()(x)
    x = Dense(10, activation='relu', name='encoded_1')(x)
    encoded = Dense(encoded_layer_size, activation='relu', name='encoded')(x)
    x = Dense(10, activation='relu', name='encoded_2')(encoded)
    # at this point the representation is (4, 4, 8) i.e. 128-dimensional
        # Decoder
    x = Dense(np.prod(shape_before_flattening[1:]),
                activation='relu')(x)
    # Reshape into an image of the same shape as before our last `Flatten` layer
    x = Reshape(shape_before_flattening[1:])(x)

    # x = UpSampling2D((2, 2))(x)
    # x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)

    x = UpSampling2D((2, 2))(x)
    x = Conv2D(IMAGE_SHAPE[0], (3, 3), activation='relu', padding='same')(x)

    x = UpSampling2D((2, 2))(x)

    decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

    autoencoder = Model(input_img, decoded)
    autoencoder.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])

    autoencoder.summary()
    encoder =  Model(inputs=input_img, outputs=encoded, name='encoder')
    return encoder,autoencoder
````

````python
def add_cluster(encoder):
    clustering_layer = ClusteringLayer(2, name='clustering')(encoder.output)
    model = Model(inputs=encoder.input, outputs=clustering_layer)
    model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
    model.summary() 
    return model
````

````python
def generate_training_data(test_size):
    x,y= sample_imgs(f['features'], f['targets'], test_size)
    x=x[:,:,8:24,8:24]
    x=x.reshape((x.shape[0],16,16,1))
    x=np.divide(x, 255.)
    return x,y
````

````python

from sklearn.cluster import KMeans
from sklearn import metrics

def fit_with_cluster(model,IS_MAX,x,y,encoder):
    ty = np.zeros((y.shape[0],2))
    print(ty.shape)
    print(y.shape)
                 
    if IS_MAX:
        v = [0, 1]
    else:
        v = [1, 0]
    ty[np.where(y==1)[0]] = v
    ty[np.where(y==0)[0]] = v[::-1]        
    kmeans = KMeans(n_clusters=2, n_init=20)
    y_pred = kmeans.fit_predict(encoder.predict(x))
    model.get_layer(name='clustering').set_weights([kmeans.cluster_centers_])    
    model.fit(x, ty, epochs=50, batch_size=10,verbose=1, shuffle=True)
    
````

````python
def draw_p(pred,pre_y):
    %matplotlib inline
    plt.figure(figsize=(8,8))
    y1 = pred[np.where(pre_y > 0)[0]]
    y2 = pred[np.where(pre_y == 0)[0]]
    plt.plot(*zip(*y1),'b.',*zip(*y2),'r.') 
    plt.show()       
def draw(model,x):
    pred = model.predict(x)
    %matplotlib inline
    plt.figure(figsize=(8,8))
    y_pred = pred.argmin(1)
    y1 = pred[np.where(y_pred > 0)[0]]
    y2 = pred[np.where(y_pred == 0)[0]]
    print(y1.shape)
    plt.plot(*zip(*y1),'b.',*zip(*y2),'r.') 
    plt.show()    
````

````python
from sklearn import metrics
def auc(y,pre_y):
    fpr, tpr, _ = metrics.roc_curve(y,pre_y)    
    auc = metrics.auc(fpr, tpr)

    plt.figure(figsize=(8,8))
    if auc< 0.5:
        fpr, tpr = tpr , fpr
        auc = 1- auc
    print("auc = %f" %auc)        
    plt.plot(fpr, tpr, label="(AUC={:.2f})".format(auc))
    plt.legend(loc='best')
    plt.show()
    dic = {}
    dic["fpr"] = fpr
    dic["tpr"] = tpr
    dic["auc"] = auc
    return dic
````

````python

def test_model(model, test_size,IS_MAX):
    x,y= sample_imgs(f['features'], f['targets'], test_size)
    x=x[:,:,8:24,8:24]
    x=x.reshape((x.shape[0],16,16,1))
    x=np.divide(x, 255.)

    q = model.predict(x)


    if IS_MAX:
        y_pred = q.argmax(1)
    else:
        y_pred = q.argmin(1)    
#     draw_p(q,y)    
    y=y.flatten()
    counter = np.where(y==y_pred)[0].shape[0]
    print("acc = %f" %(counter*1.0/len(y)))
    return auc(y,q[:,1])
    
def test_model1(model,encoder, test_size,IS_MAX):
    x,y= sample_imgs(f['features'], f['targets'], test_size)
    x=x[:,:,8:24,8:24]
    x=x.reshape((x.shape[0],16,16,1))
    x=np.divide(x, 255.)

    q = model.predict(x)
    encoded= encoder.predict(x, verbose=0)
    draw_p(encoded,y)

    if IS_MAX:
        y_pred = q.argmax(1)
    else:
        y_pred = q.argmin(1)    
    draw_p(q,y) 
    y=y.flatten()
    counter = np.where(y==y_pred)[0].shape[0]
    print("acc = %f" %(counter*1.0/len(y)))
    auc(y,q[:,1])
````
